```python
root = '/Users/rdubnic2/Desktop/ht-delete-notices/'

ht_text_pd = '/Users/rdubnic2/Desktop/ht-delete-notices/ht-text-pd/'
ht_text_pd_world = '/Users/rdubnic2/Desktop/ht-delete-notices/ht-text-pd-world/'
ht_text_pd_open_access = '/Users/rdubnic2/Desktop/ht-delete-notices/ht-text-pd-open-access/'
ht_text_pd_world_open_access = '/Users/rdubnic2/Desktop/ht-delete-notices/ht-text-pd-world-open-access/'

#file_list = os.listdir(root)
```

```python
htpd_file_list = os.listdir(ht_text_pd)
htpd_world_file_list = os.listdir(ht_text_pd_world)
htpd_oa_file_list = os.listdir(ht_text_pd_open_access)
htpd_world_oa = os.listdir(ht_text_pd_world_open_access)

# htpd_file_list
# htpd_world_file_list
# htpd_oa_file_list
# htpd_world_oa

print(f"htpd files: {len(htpd_file_list)}")
print(f"htpd_world files: {len(htpd_world_file_list)}")
print(f"htpd_oa files: {len(htpd_oa_file_list)}")
print(f"htpd_world_oa files: {len(htpd_world_oa)}")
```

```python
htpd = {}
branch = 'ht-text-pd/'

for f in htpd_file_list:
    full_path = str(root+branch+f)
    # print(full_path)
    with open(full_path, 'r') as e:
        # print(e)
        msg = email.message_from_file(e)
        if msg.is_multipart():
            for part in msg.walk():
                payload = part.get_payload(decode=True) #returns a bytes object
                strtext = payload.decode() #utf-8 is default
                dt = msg.get('date')
                # print(dt)
                # print(strtext)
                htpd[dt] = strtext

        else:
            payload = msg.get_payload(decode=True)
            strtext = payload.decode()
            dt = msg.get('date')
            # print(dt)
            # print(strtext)
            htpd[dt] = strtext
            
htpdworld = {}
branch2 = 'ht-text-pd-world/'

for f in htpd_world_file_list:
    full_path = str(root+branch2+f)
    # print(full_path)
    with open(full_path, 'r') as e:
        # print(e)
        msg = email.message_from_file(e)
        if msg.is_multipart():
            for part in msg.walk():
                payload = part.get_payload(decode=True)
                strtext = payload.decode()
                dt = msg.get('date')
                # print(dt)
                # print(strtext)
                htpdworld[dt] = strtext
        else:
            payload = msg.get_payload(decode=True)
            strtext = payload.decode()
            dt = msg.get('date')
            # print(dt)
            # print(strtext)
            htpdworld[dt] = strtext
            
htpdoa = {}
branch3 = 'ht-text-pd-open-access/'

for f in htpd_oa_file_list:
    full_path = str(root+branch3+f)
    # print(full_path)
    with open(full_path, 'r') as e:
        # print(e)
        msg = email.message_from_file(e)
        if msg.is_multipart():
            for part in msg.walk():
                payload = part.get_payload(decode=True)
                strtext = payload.decode()
                dt = msg.get('date')
                # print(dt)
                # print(strtext)
                htpdoa[dt] = strtext

        else:
            payload = msg.get_payload(decode=True)
            strtext = payload.decode()
            dt = msg.get('date')
            # print(dt)
            # print(strtext)
            htpdoa[dt] = strtext
            
htpdworldoa = {}
branch4 = 'ht-text-pd-world-open-access/'

for f in htpd_world_oa:
    full_path = str(root+branch4+f)
    # print(full_path)
    with open(full_path, 'r') as e:
        # print(e)
        msg = email.message_from_file(e)
        if msg.is_multipart():
            for part in msg.walk():
                payload = part.get_payload(decode=True)
                strtext = payload.decode()
                dt = msg.get('date')
                # print(dt)
                # print(strtext)
                htpdworldoa[dt] = strtext

        else:
            payload = msg.get_payload(decode=True)
            strtext = payload.decode()
            dt = msg.get('date')
            # print(dt)
            # print(strtext)
            htpdworldoa[dt] = strtext
```

```python
htpd_line_dict = {}

for k,v in htpd.items():
    # print(k)
    # print(v)
    lines = v.splitlines()
    # print(lines)
    for l in lines:
        # print(l)
        if k in htpd_line_dict:
            htpd_line_dict[k].append(l)
        else:
            htpd_line_dict[k] = [l]  
    
htpdworld_line_dict = {}

for k,v in htpdworld.items():
    # print(k)
    # print(v)
    lines = v.splitlines()
    # print(lines)
    for l in lines:
        # print(l)
        if k in htpdworld_line_dict:
            htpdworld_line_dict[k].append(l)
        else:
            htpdworld_line_dict[k] = [l]
    
    
htpdoa_line_dict = {}

for k,v in htpdoa.items():
    # print(k)
    # print(v)
    lines = v.splitlines()
    # print(lines)
    for l in lines:
        # print(l)
        if k in htpdoa_line_dict:
            htpdoa_line_dict[k].append(l)
        else:
            htpdoa_line_dict[k] = [l]
    
    
htpdworldoa_line_dict = {}

for k,v in htpdworldoa.items():
    # print(k)
    # print(v)
    lines = v.splitlines()
    # print(lines)
    for l in lines:
        # print(l)
        if k in htpdworldoa_line_dict:
            htpdworldoa_line_dict[k].append(l)
        else:
            htpdworldoa_line_dict[k] = [l]
```

```python
ht_text_pd_deletes = {}
# ht_text_pd_vols = []

for k,v in htpd_line_dict.items():
    # print(k)
    # print(v)
    for l in v:
        l_str = str(l)
        # print(l_str)
        if len(l_str) < 4:
            continue
        if l_str[3] == '.':
            # ht_text_pd_vols.append(l)
            if k in ht_text_pd_deletes:
                ht_text_pd_deletes[k].append(l)
            else:
                ht_text_pd_deletes[k] = [l]
        
        else:
            continue
            
ht_text_pd_world_deletes = {}

for k,v in htpdworld_line_dict.items():
    # print(k)
    # print(v)
    for l in v:
        l_str = str(l)
        # print(l_str)
        if len(l_str) < 4:
            continue
        if l_str[3] == '.':
            if k in ht_text_pd_world_deletes:
                ht_text_pd_world_deletes[k].append(l)
            else:
                ht_text_pd_world_deletes[k] = [l]
        
        else:
            continue
                    
ht_text_pd_oa_deletes = {}

for k,v in htpdoa_line_dict.items():
    # print(k)
    # print(v)
    for l in v:
        l_str = str(l)
        # print(l_str)
        if len(l_str) < 4:
            continue
        if l_str[3] == '.':
            if k in ht_text_pd_oa_deletes:
                ht_text_pd_oa_deletes[k].append(l)
            else:
                ht_text_pd_oa_deletes[k] = [l]
        
        else:
            continue
            
ht_text_pd_world_oa_deletes = {}

for k,v in htpdworldoa_line_dict.items():
    # print(k)
    # print(v)
    for l in v:
        l_str = str(l)
        # print(l_str)
        if len(l_str) < 4:
            continue
        if l_str[3] == '.':
            if k in ht_text_pd_world_oa_deletes:
                ht_text_pd_world_oa_deletes[k].append(l)
            else:
                ht_text_pd_world_oa_deletes[k] = [l]
        
        else:
            continue
```
